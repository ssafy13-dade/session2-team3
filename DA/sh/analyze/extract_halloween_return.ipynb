```python
# 10월 첫 거래일 종가
buy_price = oct_data_2001.iloc[0]['close']

# 5월 모든 거래일에 대해 수익률 계산
may_data_2002 = may_data_2002.copy()  # 경고 방지용
may_data_2002['수익률(%)'] = (may_data_2002['close'] - buy_price) / buy_price * 100

# 확인
print(may_data_2002[['close', '수익률(%)']])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("./single_stock/5930_stock.csv")
# df.index = pd.to_datetime(df.index)
df['date'] = pd.to_datetime(df['date'])
df.set_index ('date', inplace=True)
```

```python
for year in range(2001, 2024):
    print(f"🔄 {year} 처리 중...")

    oct_data = df[(df.index.year == year) & (df.index.month == 10)]
    may_data = df[(df.index.year == year + 1) & (df.index.month == 5)]

    if oct_data.empty or may_data.empty:
        continue

    rows = []
    for buy_date, buy_row in oct_data.iterrows():
        buy_price = buy_row['close']
        for sell_date, sell_row in may_data.iterrows():
            sell_price = sell_row['close']
            return_pct = (sell_price - buy_price) / buy_price * 100
            rows.append([
                year, buy_date.date(), sell_date.date(),
                round(buy_price, 2), round(sell_price, 2), round(return_pct, 2)
            ])

    df_temp = pd.DataFrame(rows, columns=[
        '매수연도', '매수일', '매도일', '매수종가', '매도종가', '수익률(%)'
    ])
    df_temp.to_csv("halloween_result1.csv", mode='a', index=False, header=(year == 2001))
```
