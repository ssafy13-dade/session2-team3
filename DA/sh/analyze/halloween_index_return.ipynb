```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("ko_indexes.csv")
# df.index = pd.to_datetime(df.index)
df1['date'] = pd.to_datetime(df1['date'])
df1.set_index ('date', inplace=True)
df1 = df1[df1['market'].isin(['KS11', 'KS200'])]


#data 넣을 리스트
results = []

for market in df1['market'].unique():

    # df = df1[market]
    df= df1[df1['market'] == market]
# 2001 ~ 2023년까지 반복
    for year in range(2001, 2024):
        oct_data = df[(df.index.year == year) & (df.index.month == 10)]
        may_data = df[(df.index.year == year + 1) & (df.index.month == 5)]

        print(oct_data)
        print(may_data)

        for buy_date, buy_row in oct_data.iterrows():
            buy_price = buy_row['close']
            for sell_date, sell_row in may_data.iterrows():
                sell_price = sell_row['close']
                return_pct = (sell_price - buy_price) / buy_price * 100
                # print(results)
                results.append({
                    '매수연도': year,
                    '매수일': buy_date.date(),
                    '매수종가': round(buy_price, 2),
                    '매도일': sell_date.date(),
                    '매도종가': round(sell_price, 2),
                    '수익률(%)': round(return_pct, 2),
                    'market': market
                })

    # 결과 정리
    df_returns = pd.DataFrame(results)
    # print(df_returns.head())
    # print(f"\n총 조합 수: {len(df_returns)}개")

    df_returns.to_csv("halloween_index_result.csv")

# 연 평균 수익률 구하기 (옵션)
# avg_return_per_year = df_returns.groupby('매수연도')['수익률(%)'].mean()


# print("\n연도별 평균 수익률:")
# print(avg_return_per_year)


```
