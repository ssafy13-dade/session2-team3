```python
import pandas as pd
```

```python
stock_df = pd.read_csv('ko_stock_prices.csv')
index_df = pd.read_csv('ko_indexes.csv')
```

```python
stock_df["code"] = stock_df["code"].astype(str).str.zfill(6)
```

```python
stock_df.info()
```

```python
index_df = index_df[index_df["market"].isin(["KS11", "KS200"])]
```

```python
print(index_df["market"].unique())
```

```python
merged_df = pd.merge(stock_df, index_df, on="date", how="left", suffixes=("", "_ks11"))
```

```python
merged_df = merged_df.dropna(subset=["market"])
```

```python
merged_df
```

```python
merged_df.to_csv("ko_index_puls_stock.csv", index=False)
```

```python
columns = ["code", "date", "close", "volume", "close_ks11", "market"]
simple_df = merged_df[columns]
```

```python
simple_df
```

```python
simple_df.to_csv("simple_stock_data.csv", index=False)
```
