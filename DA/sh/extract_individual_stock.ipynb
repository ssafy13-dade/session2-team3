```python
import pandas as pd 
import numpy as np 

```

```python
df = pd.read_csv('ko_stock_prices.csv')
```

```python
df
```

```python
df['code'].unique()
```

```python


# 'code' 열에서 유니크한 값들 중 특정 값이 있으면 새로운 DataFrame을 만듦
for value_to_check in df['code'].unique(): 
    print(value_to_check)
    new_df = df[df['code'] == value_to_check]
    # 결과 확인
    if new_df is not None:
        print(new_df)
        new_df.to_csv(f'{value_to_check}_stock.csv')
    else:
        print("새로운 DataFrame을 만들지 않았습니다.")


```
